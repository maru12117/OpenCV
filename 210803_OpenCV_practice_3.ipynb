## 트랙바 생성

```python
import cv2
import numpy as np

def change_background(value):
    
    image=np.full((500,500,3), (0,value,0), dtype=np.uint8)
    cv2.imshow("white", image)


cv2.namedWindow("white") #윈도우를 이름짓기
cv2.createTrackbar("trackbar", "white", 0,255, change_background)

cv2.imshow("white", image)
cv2.waitKey()
cv2.destroyAllWindows()
```

## 동영상 입력하여 도형을 그린 후 출력함

```python
import cv2
import numpy as np

video1=cv2.VideoCapture("thailand.mp4")

width=int(video1.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video1.get(cv2.CAP_PROP_FRAME_HEIGHT))
isWrite=False
videoWriter=cv2.VideoWriter()

while True:
    ret, frame = video1.read()
    
    if (video1.get(cv2.CAP_PROP_POS_FRAMES)== video1.get(cv2.CAP_PROP_FRAME_COUNT)):
        video1.open("thailand.mp4")
    cv2.line(frame, (100,100),(200,300), (0,0,255), 4)
    cv2.circle(frame, (300,300), 100, (255,0,0), 4)
    cv2.line(frame, (300,300),(800,100), (0,0,255), 4)
    cv2.putText(frame, "thailand", (300, 150), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,0,0), 3)
    
    cv2.imshow("VideoFrame", frame)
        
    if cv2.waitKey(33)==ord('w'): #누르면 시작
        print("비디오 저장")
        fourcc = cv2.VideoWriter_fourcc(*'XVID')
        videoWriter.open("Video.avi", fourcc, 30, (width, height),True)
        isWrite = True
        
    elif cv2.waitKey(33)==ord('s'): #누르면 종료
        videoWriter.release()
        isWrite = False
    
    elif cv2.waitKey(33) == ord('q'):
        break
        
    if isWrite==True: #isWrite가 True일때만..wrtie한다..
        videoWriter.write(frame)
        
videoWriter.release()    
video1.release()
cv2.destroyAllWindows()
```

## 동영상에 마우스 콜백함수를 입력하여 도형 그리기

```python
import cv2
import numpy as np

video2=cv2.VideoCapture("thailand.mp4")

width=int(video2.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video2.get(cv2.CAP_PROP_FRAME_HEIGHT))
isWrite=False
videoWriter=cv2.VideoWriter()

x1, y1, x2, y2= 0,0,0,0
def on_mouse(event, x, y, flags, param):
        if event==cv2.EVENT_LBUTTONDOWN:
            global x1,y1
            x1=x
            y1=y
        
        elif event==cv2.EVENT_RBUTTONDOWN:
            global x2,y2
            x2=x
            y2=y
            
while True:
    ret, frame = video2.read()
        
    if (video2.get(cv2.CAP_PROP_POS_FRAMES)== video2.get(cv2.CAP_PROP_FRAME_COUNT)):
        video2.open("thailand.mp4")
        
    cv2.line(frame, (x1,y1),(x2,y2), (0,0,255), 4)
    cv2.imshow("VideoFrame", frame)
    cv2.setMouseCallback("VideoFrame", on_mouse)
        
    if cv2.waitKey(33)==ord('w'): #누르면 시작
        print("비디오 저장")
        fourcc = cv2.VideoWriter_fourcc(*'XVID')
        videoWriter.open("Video1.avi", fourcc, 30, (width, height),True)
        isWrite = True
        
    elif cv2.waitKey(33)==ord('s'): #누르면 종료
        videoWriter.release()
        isWrite = False
    
    elif cv2.waitKey(33) == ord('q'):
        break
        
    if isWrite==True: #isWrite가 True일때만..wrtie한다..
        videoWriter.write(frame)
        
videoWriter.release()    
video2.release()
cv2.destroyAllWindows()
```

## 마우스 휠을 이용하여 트랙바 조작

```python
import cv2
import numpy as np

def onChange(value):
    global b
    b=value
    cv2.imshow("color window", createImage(b))
    
def createImage(b):
    return np.full((500,500,1), b, dtype=np.uint8)

def onMouse(event, x, y, flag, param):
    if event==cv2.EVENT_MOUSEWHEEL:
        global b
        if flag > 65536:
            if b<=255:
                b+=10
            else:b=255
        else:
            if b>0:
                b-=10
                print(b)
            else:b=0
        return onChange(b)

cv2.namedWindow("color window")
cv2.createTrackbar("color", "color window", 0,255, onChange)

cv2.setMouseCallback("color window", onMouse)
b=0
while True:
    cv2.imshow("color window", createImage(b))
    
    if cv2.waitKey(0)==ord('q'):
        break
    
cv2.destroyAllWindows()
```
